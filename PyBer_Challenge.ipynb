### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type using the groupby() function 
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# 2. Get the total drivers for each city type
total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
```

```python
#  3. Get the total amount of fares for each city type
total_fare = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare
```

```python
#  4. Get the average fare per ride for each city type by dividing the sum of all fares by the total rides
average_fare_per_ride = total_fare / total_rides
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type by dividing the sum of all fares by total drives
average_fare_per_driver = total_fare/ total_drivers
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame ({
    "Total Rides" : total_rides,
    "Total Drivers" : total_drivers,
    "Total Fares" : total_fare,
    "Average Fare per Ride" : average_fare_per_ride,
    "Average Fare per Driver" : average_fare_per_driver
})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_data_group = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
pyber_data_group.head()
#shows the total fare amount for each date and time 
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
pyber_data_group = pyber_data_group.reset_index()
pyber_data_group.head()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_data_pivot = pyber_data_group.pivot(index = "date", columns = "type", values = "fare")
pyber_data_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, 
##'2019-01-01':'2019-04-28'.
pyber_data_range = (pyber_data_pivot.index >= '2019-01-01') & (pyber_data_pivot.index <= '2019-04-28')
pyber_data_filtered = pyber_data_pivot.loc[pyber_data_range]
pyber_data_filtered
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
from datetime import datetime 
pyber_data_filtered.index = pd.to_datetime(pyber_data_filtered.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_data_filtered.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get 
##the sum of the fares for each week.
pyber_data_resampled = pyber_data_filtered.resample("W").sum()
pyber_data_resampled
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# pyber_data_resampled.plot(figsize = (20,5))
# plt.ylabel("Fare ($USD)")
# plt.title("Total Fare by City Type")

# fig,ax = plt.subplots(figsize = (20,5))
# ax.plot(pyber_data_resampled, label = "type")
# ax.set_title("Total Fare by City Type")
# ax.set_ylabel("Fare ($USD)")
# ax.legend()

pyber_data_resampled.plot(figsize = (20,5), ylabel = "Fare ($USD)", title = "Total Fare by City Type")

#show and save figure
plt.savefig("Analysis/PyBer_fare_summary.png")
plt.show()
```

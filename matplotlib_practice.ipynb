```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month 
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create a plot 
plt.plot(x_axis, y_axis)
#will display months as the x axis and total fares for each month as the y axis 
```

```python
#re-create line chart using object oriented interface method instead of MATLAB
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# different way to create the plot with ax.plt() where fig, ax = ___ is replaced with the following -- object oriented method
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# a quick way to create the plot with ax.plt()-- object oriented method
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and display it 
plt.plot(x_axis, y_axis)
#plt.show()
```

```python
# Create the plot and add a label for the legend.
#label will be added to the legend, color will be blue, marker added and line width changed to 2
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
# Add a grid.
plt.grid()
```

```python
#skill drill 5.14
#create the plot 
#Change the line color to green.
#Change the marker to a diamond-- https://matplotlib.org/stable/api/markers_api.html
#Change the line width to 2 points
#Add a legend for the city of Boston
ax = plt.axes()
ax.plot(x_axis, y_axis, color = "green", marker = "d", linewidth = 2, label = "Boston")
#add a legend
ax.legend()
#Add a title and axes labels
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_title('PyBer Fare by Month')
#Add a y-axis limit of 0-45
ax.set_ylim(0,45)
#Add grid lines
ax.grid()
```

```python
#skill drill 5.17 using object oriented approach to make a buble chart 
#Switch the axis so that the Fare($) data is on the x-axis.
#Change the color of the markers to sky blue.
#Change the size of the markers to 5 times each data point.
#Make the color 20% transparent.
#Add a black edge color to the circles.
#Make the linewidth of the circles 2 points.
#Add a legend for the city of Boston.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color = "blue", s= [i*5 for i in y_axis], alpha = .2, edgecolors = "black", linewidth = 2, label ="Boston")

#Add a title.
ax.set_title("PyBer fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Months")

#Add a limit for the x-axis data.
ax.set_xlim(0,50)

#Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

```python
#create pie charts using MATLAB approach 
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.pie(y_axis, labels = x_axis)
```

```python
#add percentages and make july pop out with explode function 
# make the chart bigger
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
